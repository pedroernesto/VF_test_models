```python
import os

models_path = '../../08_CloneMorphologies-h5v1'
file_entries = os.listdir(models_path)
morph_files = [f for f in file_entries if f.endswith((".swc", 'h5'))]

morph_clones_scale = [file_name for file_name in morph_files if any(substr in file_name for substr in ['Clone', 'Scale'])]
morph_cells = [os.path.join(models_path,file_name) for file_name in morph_files if file_name not in morph_clones_scale]
```

```python
import neurom
from neurom import viewer

for file_name in morph_cells:
    nrn = neurom.load_neuron(file_name); fig, ax = viewer.draw(nrn); fig.show() 
    
```
